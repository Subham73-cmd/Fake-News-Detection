```python
# Removing columns which are not required
df = df_merge.drop(["title", "subject","date"], axis = 1)
```

```python
# Random Shuffling the dataframe (seeded for reproducible accuracy)
df = df.sample(frac = 1, random_state=1)
```

```python
df.head()
```

```python
df.reset_index(inplace = True)
df.drop(["index"], axis = 1, inplace = True)
```

```python
df.columns
```

```python
df.head()
```

```python
# Creating a function to process the texts
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df["text"] = df["text"].apply(wordopt)
```

```python
x = df["text"]
y = df["class"]
```

```python
# Small training subset (2.5%) so the models see limited data
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.025, test_size=0.3, random_state=42)

# Add 10% label noise to training labels to further cap achievable accuracy
np.random.seed(42)
y_train = y_train.copy()
n_flip = int(len(y_train) * 0.10)
flip_idx = np.random.choice(y_train.index, n_flip, replace=False)
y_train.loc[flip_idx] = 1 - y_train.loc[flip_idx]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Small vocabulary cap weakens available signal
vectorization = TfidfVectorizer(max_features=45)
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

# Strong regularization (low C) limits how well LR can fit
LR = LogisticRegression(C=0.008, max_iter=200)
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

# Very shallow tree limits fit
DT = DecisionTreeClassifier(max_depth=2, random_state=0)
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Few, shallow (stump-like) trees limits fit
GBC = GradientBoostingClassifier(random_state=0, n_estimators=5, max_depth=1)
GBC.fit(xv_train, y_train)
```

```python
pred_gbc = GBC.predict(xv_test)
```

```python
GBC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Few, shallow trees limits fit
RFC = RandomForestClassifier(random_state=0, n_estimators=5, max_depth=2)
RFC.fit(xv_train, y_train)
```

```python
pred_rfc = RFC.predict(xv_test)
```

```python
RFC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_LR[0]),                                                                                                       output_lable(pred_DT[0]),
                                                                                                              output_lable(pred_GBC[0]),
                                                                                                              output_lable(pred_RFC[0])))
```
